```python
import json
```

```python
with open("../ans_data.json",'r') as fd:
    df=json.load(fd)
```

```python
ITR_NUM=4
```

```python
len(df)
```

```python
df['banana2']
```

```python
for curr_key, curr_elem in df.items():
    check=curr_elem['0']
    for the_key in check:
        if 'score' in the_key:
            check[the_key]=check[the_key][ITR_NUM]
```

```python
df['banana2']
```

```python
a1=[(df[key]['0']['score_acc'],key) for key in df]
a1=sorted(a1)
a2=[(df[key]['0']['score_jaccard'],key) for key in df]
a2=sorted(a2)
a3=[(df[key]['0']['score_dice'],key) for key in df]
a3=sorted(a3)
```

```python
def find_avg(arr):
    p=[x[0] for x in arr]
    return sum(p)/len(p)
```

```python
print(find_avg(a1))
a1
```

```python
print(find_avg(a2))
a2
```

```python
cnt=0
for curr_elem in a2:
    if curr_elem[0]>0.85:
        cnt+=1
```
